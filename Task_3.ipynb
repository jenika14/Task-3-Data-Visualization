```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("weather.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Precip Type'].fillna('None', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
```

```python
plt.figure(figsize=(10,5))
```

```python
plt.plot(df['Formatted Date'], df['Temperature (C)'])
plt.title("Temperature Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Temperature (C)")
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['Humidity'], bins=20)
plt.title("Humidity Distribution")
plt.xlabel("Humidity")
plt.ylabel("Frequency")
plt.show()
```

```python
df['Summary'].value_counts().head(10).plot(kind='bar')
plt.title("Top Weather Conditions")
plt.xlabel("Weather Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.scatter(df['Temperature (C)'], df['Humidity'])
plt.title("Temperature vs Humidity")
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.boxplot(x='Precip Type', y='Temperature (C)', data=df)
plt.title("Temperature by Precipitation Type")
plt.show()
```
